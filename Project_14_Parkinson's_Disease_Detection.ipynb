```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Analysis

```python
# loading the csv file data to pandas dataframe
parkinson_data = pd.read_csv('/content/parkinsons.csv')
```

Data pre processing

Separating the features & Target

```python
X = parkinson_data.drop(columns=['name','status'], axis=1)
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test Data

```python
X_train,  X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

model Evaluation

Accuracy Score

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy score of Training data : ', training_data_accuracy)
```

```python
# accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_Accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of the test data : ', test_data_Accuracy)
```

Building a Predictive System

```python
input_data = (245.51000,262.09000,231.84800,0.00235,0.000010,0.00127,0.00148,0.00380,0.01608,0.14100,0.00906,0.00977,0.01149,0.02719,0.00476,24.60200,0.467489,0.631653,-7.156076,0.127642,2.392122,0.097336)
# changing input data to a numpy array
input_data_as_numpy_array =np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)


if (prediction[0] == 0 ):
  print("The Person does not have Parkinson's disease")
else:
  print("The Person has Parkinson's ")
```
