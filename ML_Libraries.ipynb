**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0,2*np.pi, 200)
y = np.sin(x)
plt.plot(x,y)
```

```python
import numpy as np

x = np.linspace(0,2*np.pi, 200)
y = np.sin(x)
z = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x,y, label="sin")
ax.plot(x,z, label="cos")
ax.set_xlabel("Points")
ax.set_ylabel("Comp of Sin and Cos")
ax.set_title("Comparison Graph")
ax.legend()
plt.show()
```

```python
fig = plt.figure()
```

```python
fig, axs = plt.subplots(2,2)
axs[0][0].plot(x,y)
axs[0][1].plot(x,z)
axs[1][0].scatter(x,y)
axs[1][1].scatter(x,z)
```

**3D Plots**

```python
import numpy as np

n = 100
xs = np.linspace(0,1,n)
ys = np.sin(xs*6*np.pi)
zs = np.cos(xs*6*np.pi)

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
ax.plot(xs,ys,zs)
plt.show()
```

```python
data = {"apple": 10, "orange":15, 'lemon': 5,"lime": 20}
names = list(data.keys())
values = list(data.values())

fig, ax = plt.subplots(1,3,figsize=(9,3))
ax[0].bar(names,values)
ax[1].scatter(names, values)
ax[2].plot(names, values)
ax[2].tick_params(axis="x", rotation=55)
```
