Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a Pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns= house_price_dataset.feature_names)
```

```python
# Print first 5 rows of our DataFrame
house_price_dataframe.head()
```

```python
# add the target (price) column to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows & columns in the data frame
house_price_dataframe.shape
```

```python
# checking the missing values
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

1. Positive Correlation

2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# construct a heat map to understsand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.lf', annot_kws={'size':8}, cmap='Blues')
```

Splitting the data & target

```python
X = house_price_dataframe.drop(['price'], axis =1) # 1-dropping column, & 0-Dropping row
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training Data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model   Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print('R Squared Error : ',score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel("Predicted Prices")
plt.title('Actual Price vs Predicted Price')
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

#Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('R Squared Error : ', score_1)
print('Mean Absolute Error : ', score_2)
```
