**Imbalanced Dataset:**

A dataset with an unequal class distribution

```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
#first 5 rows of the  dataframe
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Determining the distribution of 2 classes
credit_card_data['Class'].value_counts()
```

This is Highly imbalanced dataset.



0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# seperating the leggit and fradulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Building a sample dataset containing similar distribution of legit and fraudulent transactions

No. of Fraudulent Transactions --->  492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
# Axis =0 will concatenate row wise
```
