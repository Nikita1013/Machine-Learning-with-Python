Importing the Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from  sklearn.metrics import accuracy_score
```

Data Collection & Pre-Processing

```python
# loading the dataset to pandas DataFrame
loan_dataset = pd.read_csv('/content/Loan-Prediction.csv')
```

```python
# number of missing values in each column
loan_dataset.isnull().sum()
```

```python
# dropping the missing values
loan_dataset = loan_dataset.dropna()
```

```python
#label encodeing
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}}, inplace=True)
```

```python
# Dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
# replacicng the value of 3+ to 4
loan_dataset = loan_dataset.replace(to_replace='3+', value=4)
```

```python
# convert categorical column to numerical values
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0}, 'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}}, inplace=True)
```

```python
loan_dataset.head()
```

```python
# separating the data & the label
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'])
Y = loan_dataset['Loan_Status']
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.1,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the model

Support Vector Machine

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine model
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)


print('Accuracy on training data : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)


print('Accuracy on test data : ', test_data_accuracy)
```
