Matplotlib


* Useful for making plots (making graphs and all other visualizations)



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for out plots
import numpy as np
```

```python
x = np.linspace(0,10,100)   # it gives evenly spaced 100 values between 1 to 10
y = np.sin(x)  #gives sin values of the 100 numbers
z = np.cos(x)  #gives cos values of the 100 numbers
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the Data

```python
# Sine wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine Wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title x-axis & y-axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

PLotting a Parabola

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#Using other symbols and colors to plot the diagrams
plt.plot(x,y, 'r+')   # r+ ==> r = red color and + the symbol used to draw the diagram
plt.show()
```

```python
plt.plot(x,y,'g*')   # g* ==> g= green color and * is the symbol used to plot
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

Plotting multiple graphs in same

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), 'g-')
plt.plot(x,np.cos(x), 'r.')
plt.show()
```

BAR PLOT

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])   # 0, 0 is th origin  & 1,1, height and width
languages = ['English','French','Hindi','Marathi','Korean']
people = [100,50,150,40,60]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([10,10,1,1])
languages = ['Marathi','Hindi','Korean','English','French']
people = [50,70,30,45,20]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

PIE CHART

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Korean','Hindi','Marathi']
people = [75,35,30,80,60]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(200)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
